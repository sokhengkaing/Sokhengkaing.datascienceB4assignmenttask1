## 3. Write a prgram to generate the following 3x8 matrix The object x<-1:8

### 1. cbind()

```r
c1 <- c(1, 8, 3)
c2 <- c(2, 7, 3)
c3 <- c(3, 6, 3)
c4 <- c(4, 5, 3)
c5 <- c(5, 4, 3)
c6 <- c(6, 3, 3)
c7 <- c(7, 2, 3)
c8 <- c(8, 1, 3)
cbind_matrix <- matrix(cbind(c1, c2, c3, c4, c5, c6, c7, c8), 3)
cbind_matrix
```

```r
r1 <- c(1, 2, 3, 4, 5, 6, 7, 8)
r2 <- c(8, 7, 6, 5, 4, 3, 2, 1)
r3 <- c(3, 3, 3, 3, 3, 3, 3, 3)
rbind_matrix <- matrix(rbind(r1, r2, r3), 3)
rbind_matrix
```

## 11.Write a program to get the weight and height from the user to calculate the following cases
a. If your BMI is less than 18.5, it falls within the underweight range.
b. If your BMI is 18.5 to <25, it falls within the normal.
c. If your BMI is 25.0 to <30, it falls within the overweight range.
d. If your BMI is 30.0 or higher, it falls within the obese range.

```r
calcBMI <- function(weight, height) {
  return(round(weight / (height * height), 0))
}

weight <- readline(prompt = "enter your weight (kg): ")
height <- readline(prompt = "enter your height (m): ")

weight <- as.double(weight)
height <- as.double(height)

bmi = calcBMI(weight = weight, height = height)

if (bmi < 18.5) {
    print(paste('BMI: ', bmi, ' - you are under weight'))
} else if (18.5 < bmi && bmi < 25) {
    print(paste('BMI: ', bmi, ' - you are normal'))
} else if (25 < bmi && bmi < 30) {
    print(paste('BMI: ', bmi, ' - you are over weight'))
} else if (30 < bmi) {
    print(paste('BMI: ', bmi, ' - you are obese'))
}
```
